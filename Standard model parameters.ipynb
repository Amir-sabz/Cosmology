```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy.misc import derivative
import plotly.express as px
from scipy import optimize

b = 0.0493
dm = 0.264
de = 1-dm-b
c = 3*10**8
h0 = 0.674
H0 = 100*h0/(3.09*10**19)

def e(z):
  return (((b+dm)*((1+z)**3))+de)**-0.5

Z = []
Khi = []
Dl = []
Da = []
H = []
Q = []
dKhi = []
dDl = []
dDa = []
dH = []
dQ = []

def khi(z):
  return integrate.quad(e,0,z)[0]*c/(H0*3.09*10**22)
def dl(z):
  return khi(z)*(1+z)
def da(z):
  return khi(z)/(1+z)
def h(z):
  return H0*e(z)**-1*3.09*10**19
def q(z):
  return (((b+dm)*(1+z)**3)+de*(-2))*e(z)**2/2
def dkhi(z):
  return derivative(khi, z, dx=0.000001)
def ddl(z):
  return derivative(dl, z, dx=0.000001)
def dda(z):
  return derivative(da, z, dx=0.000001)
def dh(z):
  return derivative(h, z, dx=0.000001)
def dq(z):
  return derivative(q, z, dx=0.000001)

Z = np.arange(0,1100.1,0.1)
for z in Z:
  Khi.append(khi(z))
  Dl.append(dl(z))
  Da.append(da(z))
  H.append(h(z))
  Q.append(q(z))
  dKhi.append(dkhi(z))
  dDl.append(ddl(z))
  dDa.append(dda(z))
  dH.append(dh(z))
  dQ.append(dq(z))

npZ = np.array(Z)
npKhi = np.array(Khi)
npDl = np.array(Dl)
npDa = np.array(Da)
npH = np.array(H)
npQ = np.array(Q)
npdKhi = np.array(dKhi)
npdDl = np.array(dDl)
npdDa = np.array(dDa)
npdH = np.array(dH)
npdQ = np.array(dQ)

fig = px.line(x=npZ, y=npH)
fig.show()
```
